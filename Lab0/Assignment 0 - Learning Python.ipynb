now develop a 2-var function to determine BMI given height and weight. The formula for the BMI is (weight*0.45)/(height*0.025)^2 where weight is in lbs and height in inches.

```python
def BMI(height, weight):
    w = weight*0.45
    h = height*0.025
    return w/h**2

BMI(66,170)
```

now let us develop a list of weights corresponding to heights given above and compute the BMI for each pair using a map function. The zip function allows us to create a list of tuples (one from height list and one from weights list). Note that we also use the list function to display the results as a list

```python
weights = [145, 260, 172, 165, 174, 184]
tuples = list(zip(heights, weights))
```

```python
def map_tuple(func, tup):
  
    length = len(tup)
    count = 0
    new_tuple = ()
    while count < length:
        new_tuple = new_tuple +  (func(tup[count][0],tup[count][1]),) 
        count = count + 1
    return new_tuple
```

```python
list(map_tuple(BMI,tuples))
```

Next we will learn about 3 useful functions called, map, reduce and filter. These are powerful functions that allows you to accomplish interesting tasks with just a few lines of code. Let us take a look at the map function first. The syntax of the map function is that it takes two arguments, function and a list and apply the function to each element in the list. We did something really similar with the function map_tuple above, but this allows us to just keep things really simple. 
We will look at the example where we convert the heights in inches to heights in centimeters by applying a simple conversion formula

```python
def toCM(height):
    return height*2.54
list(map(toCM,heights))
    
```

The reduce function is a powerful function that iteratively applies a function to a result. For example


```python
import functools
def add(x, y):
    return x + y
functools.reduce(add,heights)
```

```python
def max(x, y):
    if x > y:
        return x
    else:
       return y 
max(4,7)
```

```python
functools.reduce(max,heights)
```
